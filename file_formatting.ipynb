```python
import csv
```

```python
kernels_to_convert = []
with open('kernel_quantiles.csv', 'r') as infile:
    for line in infile:
        line = line.lower()
        line = line.replace('.csv', '')
        line = line.replace('_comb', '')
        line = line.replace(' ', '_')
        kernels_to_convert.append(line)
```

```python
kernels_to_convert
```

```python
split_list = []
for line in kernels_to_convert:
    list_ = line.split(',')
    split_list.append(list_)
```

```python
header = split_list[0]
```

```python
rows = split_list[1:]
```

```python
# Writing to a CSV file
with open('kernels_clean.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)     # Write header
    writer.writerows(rows)      # Write data rows
```
